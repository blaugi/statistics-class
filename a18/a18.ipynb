# Análise de Correlação e Diagramas de Dispersão

```python
import pandas as pd
import itertools
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregar o dataset
file_path = 'dataset.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
# Selecionar apenas as colunas numéricas relevantes
num_cols = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
df_num = df[num_cols]
```

```python
# Calcular o coeficiente de correlação para cada par de variáveis numéricas
correlations = {}
for col1, col2 in itertools.combinations(num_cols, 2):
    corr = df_num[col1].corr(df_num[col2])
    correlations[(col1, col2)] = corr
correlations
```

```python
# Plotar diagramas de dispersão para cada par de variáveis numéricas
for col1, col2 in itertools.combinations(num_cols, 2):
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=df_num[col1], y=df_num[col2])
    plt.title(f'Diagrama de Dispersão: {col1} vs {col2}')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.show()
```

## Análise dos Resultados

- Coeficientes de correlação próximos de 1 ou -1 indicam forte relação linear positiva ou negativa, respectivamente.
- Coeficientes próximos de 0 indicam pouca ou nenhuma relação linear.
- Os diagramas de dispersão permitem visualizar se a relação é realmente linear ou se há padrões não-lineares.
- Compare os valores dos coeficientes com os gráficos para validar se a relação visual condiz com o valor numérico.
