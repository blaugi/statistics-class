```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
```

```python
df = pd.read_csv('dataset.csv', index_col='Rank')
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
sales_per_year = df.groupby('Year')[['Global_Sales', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum().reset_index()
sales_per_year
```

### 1. Gr√°fico em linha

```python
plt.figure(figsize=(10, 6))
plt.plot(sales_per_year['Year'], sales_per_year['Global_Sales'], linestyle='-', color='red', label='Global Sales')

plt.legend()
plt.title(f"Global sales of video games by year")
plt.xlabel("Year")
plt.ylabel("Global Sales (millions of units)")
plt.grid()
plt.show()
```

### An√°lise:
- Podemos ver que h√° um crescimento no n√∫mero de jogos vendidos de acordo com os anos e depois de 2010 o dataset bem possivelmente n√£o manteve todos os jogos que foram lan√ßados.


## 2. Gr√°fico de √Årea

```python
plt.figure(figsize=(8, 5))
alpha=0.4
plt.fill_between(sales_per_year['Year'], sales_per_year['Global_Sales'], linestyle='-', color='red', label='Global Sales', alpha=alpha)
plt.fill_between(sales_per_year['Year'], sales_per_year['EU_Sales'], linestyle='-', color='blue', label='EU Sales', alpha=alpha)
plt.fill_between(sales_per_year['Year'], sales_per_year['NA_Sales'], linestyle='-', color='yellow', label='NA Sales', alpha=alpha)
plt.fill_between(sales_per_year['Year'], sales_per_year['JP_Sales'], linestyle='-', color='green', label='JP Sales', alpha=alpha)
plt.fill_between(sales_per_year['Year'], sales_per_year['Other_Sales'], linestyle='-', color='cyan', label='Other Sales', alpha=alpha)
plt.legend()
plt.title(f"Area graph")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid() 
plt.show()
```

## An√°lise
- Podemos perceber que h√° um padr√£o de crescimento que se mant√©m nas diferentes regi√µes exceto o Jap√£o que parece se manter.

## 5. Gr√°fico de Dispers√£o

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['Global_Sales'], s=df['Global_Sales']*10, c='red', alpha=0.6)
plt.title('Global Sales by Year (Scatter Chart)')
plt.xlabel('Release Year')
plt.ylabel('Global Sales (millions)')
plt.grid(True)
plt.show()
```

### An√°lise 
- Percebe-se um padr√£o semelhante ao do gr√°fico de linha mas evidenciando o qu√£o incomum foram as vendas entre 2005 e 2010

## 6. Gr√°fico de Caixa

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='NA_Sales', by='Genre', grid=False)
plt.title('Distribution of North America Sales by Genre (Box Chart)')
plt.suptitle('')  # Remove the default pandas title
plt.xlabel('Genre')
plt.ylabel('NA Sales (millions)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### An√°lise
- Nesse caso o boxplot n√£o trouxe muitras informa√ß√µes interessantes dado a distribui√ß√£o dos dados. Testamos com outras variav√©is sem sucesso

## Gr√°fico de Torta

```python
plt.figure(figsize=(15, 15))
platform_sales = df.groupby('Platform')['Global_Sales'].sum()
plt.pie(platform_sales, labels=platform_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Global Sales Distribution by Platform (Pie Chart)')
plt.show()
```
